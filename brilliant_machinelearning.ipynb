```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

song_lyrics = pd.read_csv('song_lyrics.csv')
song_lyrics.head()
```

```python
song_lyrics = song_lyrics.where(song_lyrics['language'] == 'en')
song_lyrics = song_lyrics.dropna()
song_lyrics = song_lyrics.drop(['language'], axis=1)
song_lyrics = song_lyrics.drop(['language_cld3'], axis=1)
song_lyrics = song_lyrics.drop(['language_ft'], axis=1)
song_lyrics.head()
```

```python
song_lyrics['lyrics'] = song_lyrics['lyrics'].str.replace('\n', ' ').str.replace(r"\[.*?\]", "")
```

```python
print(song_lyrics['lyrics'][0])
```

```python
song_lyrics[0:1000].to_csv('song_lyrics_1000.csv', index=False)
```
